```python
df = df[['G1','G2','G3','absences','activities','paid','failures','studytime']]
df.head()
```

```python
X = df.drop(['G3'],axis=1)
Y = df['G3']
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X = pd.DataFrame(scaler.fit_transform(X),columns=X.columns)
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size=0.25,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train,y_train)
y_pred = lr.predict(X_test)
print(r2_score(y_test,y_pred))
print("Cross val score : ",cross_val_score(lr,X,Y,cv=5).mean())
```

```python
print(mean_absolute_error(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
print(np.sqrt(mean_squared_error(y_test,y_pred)))
```

```python
dt = DecisionTreeRegressor()
dt.fit(X_train,y_train)
y_dt = dt.predict(X_test)
print(r2_score(y_test,y_dt))
print("Cross val score : ",cross_val_score(dt,X,Y,cv=5).mean())
```

```python
print(mean_absolute_error(y_test,y_dt))
print(mean_squared_error(y_test,y_dt))
print(np.sqrt(mean_squared_error(y_test,y_dt)))
```

```python
rf = RandomForestRegressor()
rf.fit(X_train,y_train)
y_rf = rf.predict(X_test)
print(r2_score(y_test,y_rf))
print("Cross val score : ",cross_val_score(rf,X,Y,cv=5).mean())
```

```python
print(mean_absolute_error(y_test,y_rf))
print(mean_squared_error(y_test,y_rf))
print(np.sqrt(mean_squared_error(y_test,y_rf)))
```
